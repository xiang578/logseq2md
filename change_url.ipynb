```python

file = "url_title.txt"
cnt = 0
out = open("out.html", "w")
out.write('<META HTTP-EQUIV="Content-Type" CONTENT="text/html; charset=UTF-8">\n')
out.write('<Title>Bookmarks</Title>\n')
out.write('<H1>Bookmarks</H1>\n')
out.write('<DL><p>\n')

for line in open(file):
    if "?p=" not in line or "comment" in line or "#more" in line:
        continue
    # print(line.split())
    url, title = line.rstrip().split(' ',1)
    out.write('<DT><A HREF="{}">{}</A></DT>\n'.format(url, title))
    cnt += 1
    # if cnt > 5:
        # break
out.write('</DL><p>\n')
print(cnt)
```
